Exploratory Data Analysis (EDA)

This notebook explores the cleaned Life Expectancy dataset to identify patterns, trends, and relationships between life expectancy and various socio-economic and health-related factors.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

df = pd.read_csv('../data/life_expectancy_cleaned.csv')
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Life expectancy '], kde=True)
plt.title('Distribution of Life Expectancy')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='GDP', y='Life expectancy ', data=df)
plt.title('Life Expectancy vs GDP')
plt.show()
```
