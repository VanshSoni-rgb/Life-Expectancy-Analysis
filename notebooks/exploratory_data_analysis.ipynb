Exploratory Data Analysis (EDA)

This notebook explores the cleaned Life Expectancy dataset to identify patterns, trends, and relationships between life expectancy and various socio-economic and health-related factors.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

df = pd.read_csv('../data/life_expectancy_cleaned.csv')
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Life expectancy '], kde=True)
plt.title('Distribution of Life Expectancy')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='GDP', y='Life expectancy ', data=df)
plt.title('Life Expectancy vs GDP')
plt.show()
```

```python
plt.figure(figsize=(12,10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm', annot=False)
# After you have drawn your heatmap
plt.tight_layout()  
plt.savefig("../images/correlation_heatmap.png", bbox_inches='tight')
plt.show() 

```

Correlation Analysis

The correlation heatmap indicates that life expectancy is strongly positively correlated with schooling, income composition of resources, and GDP. Adult mortality shows a strong negative correlation with life expectancy, highlighting its significant impact on population health.


```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Schooling', y='Life expectancy ', data=df)
plt.title('Life Expectancy vs Schooling')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Status', y='Life expectancy ', data=df)
plt.title('Life Expectancy by Country Status')
plt.show()
```

Key Insights

 Higher schooling levels and income composition are associated with increased life expectancy.
 Countries with higher adult mortality tend to have lower life expectancy.
 Developed countries generally exhibit higher and more stable life expectancy compared to developing countries.
 Economic indicators such as GDP show a positive relationship with life expectancy.


Limitations

 Correlation does not imply causation.
 Mean imputation and outlier treatment may slightly smooth extreme real-world variations.
 The dataset does not include recent global health crises.


Conclusion

This project demonstrates how socio-economic and health-related factors influence life expectancy across countries. Through systematic data cleaning and exploratory analysis, key indicators affecting population health were identified, providing valuable insights for policy and decision-making.
